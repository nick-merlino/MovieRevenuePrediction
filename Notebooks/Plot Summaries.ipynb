```python
def NBPredict(test_summary,log_prior,log_likelihood,classes,vocab):
    test_lemmas = []
    for word,tag in nltk.pos_tag(test_summary):
        if word.lower() not in stop_list and word.isalpha():
            lemma = wordnet_lemmatizer.lemmatize(word,pos=get_wordnet_pos(tag))
            test_lemmas.append(lemma)
            
    for c in classes:
        total[c] = logprior[c]
        for word in test_lemmas:
            if word in vocab:
                total[c] += log_likelihood[(word,c)]
    return max(total, key=total.get)
```

```python
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return 'n'
```
